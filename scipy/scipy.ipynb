integraion functions---> quad function calculates integration of one variable, dblquad function calculates the integration of two variables

```python
from scipy import integrate
help(integrate)
```

```python
i=scipy.integrate.quad(lambda x:special.exp10(x),0,1)
print(i)
```

```python
e=lambda x,y : x**10 * y**x
l= lambda x : 1
u =lambda x :-1
integrate.dblquad(e,0,2,l,u)
```
